```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

# Load the dataset
df = pd.read_csv('cleaned_spam_dataset.csv')

class_distribution = df['label'].value_counts()

# Plot a bar chart for email text classes
class_distribution.plot(kind='bar', rot=0)
plt.title('Distribution of Email Text Classes')
plt.xlabel('Email Text Class')
plt.ylabel('Count')
plt.show()

# Add a new column for text length
df['text_length'] = df['text'].apply(len)

# Set a limit for the x-axis range
x_axis_limit = 1000

# Plot a histogram for email text length
plt.hist(df['text_length'], bins=50, edgecolor='k', range=(0, x_axis_limit))
plt.title('Distribution of Email Text Length')
plt.xlabel('Email Text Length')
plt.ylabel('Count')
plt.show()



```
